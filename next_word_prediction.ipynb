```python
import numpy as np
import tensorflow as tf
```

```python
with open('/content/python.txt', 'r', encoding='utf-8') as file:
    python_text = file.read()
```

```python
python_text
```

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
```

```python
# tokenized extracted text using tokenizer
tokenizer.fit_on_texts([python_text])
```

```python
tokenizer.get_config()
```

```python
tokenizer.word_index
```

```python
input_sentences = []
for sentence in python_text.split('\n'):
  print(sentence)
  token_list = tokenizer.texts_to_sequences([sentence])[0]
  print(token_list)
  for i in range(1, len(token_list)):
    sequence = token_list[:i+1]
    input_sentences.append(sequence)
```

```python
input_sentences[:5] # see there not in the same length so we want to add paddings to each
```

```python
max_length = max([len(input_sentence) for input_sentence in input_sentences])
max_length
```

```python
input_sentences = np.array(tf.keras.preprocessing.sequence.pad_sequences(input_sentences, maxlen=max_length, padding='pre'))
```

```python
input_sentences[0] # here last -> label , and rest are -> features
```

```python
X = input_sentences[:, :-1] # features
y = input_sentences[:, -1] # label
```

```python
X[0], y[0]
```

```python
no_of_classes = len(tokenizer.word_index) + 1 # here no of classes means by -> how many numbers that we have in this particular dataset
no_of_classes
```

```python
# make our labels in order to no_of_classes (onehot encoding)
y = np.array(tf.keras.utils.to_categorical(y, num_classes=no_of_classes))
```

```python
y[0]
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Embedding(no_of_classes, 80, input_length=max_length-1))
model.add(tf.keras.layers.LSTM(100))
model.add(tf.keras.layers.Dense(no_of_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
history = model.fit(X, y, epochs=100)
```

```python
input_text = "Python is"
token_list = tokenizer.texts_to_sequences([input_text])[0]
print(token_list)
```

```python
token_list = tf.keras.preprocessing.sequence.pad_sequences([token_list], maxlen=max_length-1, padding='pre')
token_list
```

```python
preds = np.argmax(model.predict(token_list), axis=-1)
preds
```

```python
for word, index in tokenizer.word_index.items():
  if index == preds:
    print(word)
```

```python
def print_pred_word(input: str, no_of_words_to_pred: int):
  for i in range(no_of_words_to_pred):
    token_list = tokenizer.texts_to_sequences([input])[0]
    token_list = tf.keras.preprocessing.sequence.pad_sequences([token_list], maxlen=max_length-1, padding='pre') # add padding
    preds = np.argmax(model.predict(token_list), axis=-1)
    output = ""
    for word, index in tokenizer.word_index.items():
      if index == preds:
        output = word
        break
    input += " " + output
  return input
```

```python
print(print_pred_word("python is a high level general purpose programming language", 100))
```
